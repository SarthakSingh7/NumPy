## Important example: use of -1 in shape selection

 let us assuse that we have have a big array like array like n15
 now we want that the shape of the array n15 should be 5 rows and we dont know the number of columns.

 so there is a way to fix the number of rows and the number of columns will automaticlally be adjusted
 to do that we just need to fix the shape as 
 

```
n15.shape = (5,-1)
```


 when **-1** is used in place of columns that implied that the **program will automatically select** the columns and the
 rows will be fixed to 5
 

````python
n15.shape = (5,-1)
n15
````

````python
n15.shape = (-1,10)
n15
````

````python
n15.shape = (-1,6)
n15

# this is how we can automatically asjust the shape of a numpy array
````

## set difference 

````python
n19 = np.array([10,20,30,40,50,60])
n20 = np.array([50,60,70,80,90])

# this is n19 - n20
np.setdiff1d(n19,n20)

# we can say that these elements are unique in n19
````

# <div align = 'center'> **NumPy Array Mathematics** 

# Addition of NumPy arrays

````python
n21 = np.array([10,20])
n22 = np.array([30,40])

# 10 + 20 + 30 + 40 

np.sum([n21,n22])
````

````python
np.sum([n21,n22],axis = 0)

# adds with respect to column

# [10 20]
#  +   +
# [30 40] 
#    =
# [40 60] 
````

````python
np.sum([n21,n22],axis = 1)

# addition with respect to rows

# [10 20] implies 10 + 20 = 30
# [30 40] implies 30 + 40 = 70
# final = [30,70]
````
